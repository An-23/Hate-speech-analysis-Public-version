```python
import pandas as pd
```

## For Val data

```python
bert=pd.read_csv('./Metrics/Valmet/met_Bert_val_hasoc.csv')
indic=pd.read_csv('./Metrics/Valmet/met_Indic_val_hostile.csv')
muril=pd.read_csv('./Metrics/Valmet/met_Muril_val_final.csv')
robetaciv=pd.read_csv('./Metrics/Valmet/met_Roberta_val_civil.csv')
roberta=pd.read_csv('./Metrics/Valmet/met_Roberta_val_final.csv')
```

```python
# Extract metrics
bert_metrics = bert.iloc[0:5,1]
indic_metrics = indic.iloc[0:5,1]
muril_metrics = muril.iloc[0:5,1]
robertaciv_metrics = robetaciv.iloc[0:5,1]
roberta_metrics = roberta.iloc[0:5,1]

# metrics = bert.columns[1:].tolist()
models = ['BERT', 'Indic', 'Muril', 'RobertaCiv', 'Roberta'] # Manually define model names

# Create the dataframe
df = pd.DataFrame({
    'Model Name': models,
    'Accuracy': [bert_metrics[0], indic_metrics[0], muril_metrics[0], robertaciv_metrics[0], roberta_metrics[0]],
    'F1-score': [bert_metrics[1], indic_metrics[1], muril_metrics[1], robertaciv_metrics[1], roberta_metrics[1]],
    'Precision': [bert_metrics[2], indic_metrics[2], muril_metrics[2], robertaciv_metrics[2], roberta_metrics[2]],
    'Recall': [bert_metrics[3], indic_metrics[3], muril_metrics[3], robertaciv_metrics[3], roberta_metrics[3]],
    'AUROC': [bert_metrics[4], indic_metrics[4], muril_metrics[4], robertaciv_metrics[4], roberta_metrics[4]]
})

# Set 'Model' column as index
df.set_index('Model Name', inplace=True)

df
```

```python
# prompt: Using dataframe df: Highlight the max in each column

def highlight_max(s):
    '''
    highlight the maximum in a Series Bold.
    '''
    is_max = s == s.max()
    return ['font-weight: bold' if v else '' for v in is_max]

df.style.apply(highlight_max)
```
